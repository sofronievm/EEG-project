```python
import pyedflib
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import math
import pandas as pd
import scipy.stats as scp
import time
import sys

start = time.time()

path = 'G:/EEG Project/DREAMS database/data raw/'

for idx, filename in enumerate(glob.glob(os.path.join(path, '*.edf'))):
    start2 = time.time()

    f = pyedflib.EdfReader(filename)
    
    n = f.signals_in_file
    signal_labels = f.getSignalLabels()
    n_min = f.getNSamples()[0]
    sigbufs = [np.zeros(f.getNSamples()[i]) for i in np.arange(n)]
    for i in np.arange(n):
        sigbufs[i] = f.readSignal(i)
        if n_min < len(sigbufs[i]):
            n_min = len(sigbufs[i])
    f._close()
    del f
    ECG = sigbufs[0]
    ch1 = sigbufs[1]
    ch2 = sigbufs[2]
    EMG1 = sigbufs[3]
    EOG1 = sigbufs[4]
    Vth = sigbufs[5]
    Vab = sigbufs[6]
    Naf2p_A1 = sigbufs[7]
    Naf1 = sigbufs[8]
    PHONO = sigbufs[9]
    PR = sigbufs[10]
    SAO2 = sigbufs[11]
    PCPAP = sigbufs[12]
    POS = sigbufs[13]
    EOG2 = sigbufs[14]
    ch3 = sigbufs[15]
    ch4 = sigbufs[16]
    ch5 = sigbufs[17]
    ch6 = sigbufs[18]
    EMG2 = sigbufs[19]
    PULSE = sigbufs[20]
    VTOT = sigbufs[21]
    EMG3 = sigbufs[22]
    single_channel = np.subtract(ch5, ch6)
    n = 1000
    size = int(len(single_channel)/n)
    SCEEG_epochs = np.zeros((size, n))
    SCEEG_epochs = [single_channel[i:i + n] for i in range(0, len(single_channel), n)]
    epochs_to_csv = pd.DataFrame(SCEEG_epochs)
    epochs_to_csv = epochs_to_csv.transpose()
    epochs_to_csv.to_csv(filename+'-epochs_ch5-ch6.csv')
    end2 = time.time()
    elapsed2 = end2 - start2
    print("processed document :", filename)
    print("Time for document: %.2f" % elapsed2, "secs")
end = time.time()
elapsed = (end - start)/60
print("Time to complete set: %.2f " % elapsed, "mins")
```

```python
import pyedflib
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import math
import pandas as pd
import scipy.stats as scp

path = 'data/'

for idx, filename in enumerate(glob.glob(os.path.join(path, 'EEG*/*.edf'))):

    f = pyedflib.EdfReader(filename)
    
    n = f.signals_in_file
    signal_labels = f.getSignalLabels()
    n_min = f.getNSamples()[0]
    sigbufs = [np.zeros(f.getNSamples()[i]) for i in np.arange(n)]
    for i in np.arange(n):
        sigbufs[i] = f.readSignal(i)
        if n_min < len(sigbufs[i]):
            n_min = len(sigbufs[i])
    f._close()
    del f
    EOGL = sigbufs[0]
    EOGR = sigbufs[1]
    EMR = sigbufs[2]
    Oz = sigbufs[3]
    Cz = sigbufs[4]
    Fpz = sigbufs[5]
    C3 = sigbufs[6]
    C4 = sigbufs[7]
    single_channel = np.subtract(Oz, C3)
    n = 3840
    size = math.ceil(len(single_channel)/n)
    SCEEG_epochs = np.zeros((size, n))
    SCEEG_epochs = [single_channel[i:i + n] for i in range(0, len(single_channel), n)]
    columns = []
    for i in range(size):
        columns.append(str(i+1))
    # print(columns)
    
    epochs_to_csv = pd.DataFrame(SCEEG_epochs)
    epochs_to_csv = epochs_to_csv.transpose()
    epochs_to_csv.to_csv('CSV_epoch_data/EEG-'+str(idx+1)+'-epochs.csv')
#     EEG101.describe().to_csv('data/EEG data 3/EEG3'+str(idx+1)+'_stats.csv')
# #     print("writing file EEG1" +str(idx+1)+ "_stats.csv")
#     kurtosis_list = []
#     skewness_list = []
#     zero_crossing_rate_list = []
#     for n in range(size):
#         kurtosis_list.append(scp.kurtosis(SCEEG_epochs[n]))
#         skewness_list.append(scp.skew(SCEEG_epochs[n]))
#         zero_crossing_rate_list.append(((SCEEG_epochs[n][:-1] * SCEEG_epochs[n][1:]) < 0).sum())
#     # print(kurtosis_list)
#     # print(skewness_list)
#     kurt = pd.DataFrame([kurtosis_list])
#     skew = pd.DataFrame([skewness_list])
#     ZCR = pd.DataFrame([zero_crossing_rate_list])
#     score_file = pd.read_csv("data/Hypnograms/30sec hypnograms/EEG data 3/EEG3"+str(idx+1)+".csv")
#     scores = score_file['Score']
#     EEG_stats = pd.read_csv('data/EEG data 3/EEG3'+str(idx+1)+'_stats.csv', index_col=0)
#     EEG101.describe().append(kurt).append(skew).append(ZCR).append(scores).to_csv('data/EEG data 3/EEG3'+str(idx+1)+'_stats.csv')
#     EEG_stats = pd.read_csv('data/EEG data 3/EEG3'+str(idx+1)+'_stats.csv', index_col=0)
#     EEG_stats = EEG_stats.transpose()
#     EEG_stats.columns = ['count','mean','std','min', '25%','50%','75%','max','kurtosis','skewness', 'ZCR', 'scores']
#     EEG_stats.to_csv('time_features_Oz-Cz/EEG3'+str(idx+1)+'_features.csv')
    print("processing document " +str(idx+1)+' ...')
print("Completed!")
```

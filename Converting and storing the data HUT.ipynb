```python
import pyedflib
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import math
import pandas as pd
import scipy.stats as scp
import sys
import time
start = time.time()

path = 'G:/EEG Project/Hut lab data/30_sec_epochs/data_raw/'

for idx, filename in enumerate(glob.glob(os.path.join(path, '*.edf'))):
    start2 = time.time()

    f = pyedflib.EdfReader(filename)
    
    n = f.signals_in_file
    signal_labels = f.getSignalLabels()
    n_min = f.getNSamples()[0]
    sigbufs = [np.zeros(f.getNSamples()[i]) for i in np.arange(n)]
    for i in np.arange(n):
        sigbufs[i] = f.readSignal(i)
        if n_min < len(sigbufs[i]):
            n_min = len(sigbufs[i])
    f._close()
    del f
    EOGL = sigbufs[0]
    EOGR = sigbufs[1]
    EMR = sigbufs[2]
    Oz = sigbufs[3]
    Cz = sigbufs[4]
    Fpz = sigbufs[5]
    C3 = sigbufs[6]
    C4 = sigbufs[7]
    single_channel = np.subtract(C3, C4)
    n = 3840      #1280 for the 10 second epochs
    size = int(len(single_channel)/n)
    SCEEG_epochs = np.zeros((size, n))
    SCEEG_epochs = [single_channel[i:i + n] for i in range(0, len(single_channel), n)]
    columns = []
    for i in range(size):
        columns.append(str(i+1))
    epochs_to_csv = pd.DataFrame(SCEEG_epochs)
    epochs_to_csv = epochs_to_csv.transpose()
#     to_print = filename.split('data_raw\', 1)[1]
    epochs_to_csv.to_csv(filename+'-epochs_C3-C4.csv')
    end2 = time.time()
    elapsed2 = end2 - start2
    print("processed document :", filename)
    print("Time for document: %.2f" % elapsed2, "secs")
end = time.time()
elapsed = (end - start)/60
print("Time to complete set: %.2f " % elapsed, "mins")
```
